```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load your dataset (replace 'data.csv' with your dataset file)
data = pd.read_csv('crime.csv')

# Extract the 'PctPopUnderPov' and 'ViolentCrimesPerPop' columns
pct_pop_under_pov = data['PctPopUnderPov'].values.reshape(-1, 1)
violent_crimes_per_pop = data['ViolentCrimesPerPop'].values.reshape(-1, 1)

# Create a scatter plot of the original data
plt.figure(figsize=(10, 5))
plt.scatter(pct_pop_under_pov, violent_crimes_per_pop, alpha=0.5)
plt.title('Scatter Plot of Original Data')
plt.xlabel('PctPopUnderPov')
plt.ylabel('ViolentCrimesPerPop')
plt.show()

# Create a Min-Max scaler
scaler = MinMaxScaler()

# Scale the features
scaled_pct_pop_under_pov = scaler.fit_transform(pct_pop_under_pov)
scaled_violent_crimes_per_pop = scaler.fit_transform(violent_crimes_per_pop)

# Create a scatter plot with the scaled values
plt.figure(figsize=(10, 5))
plt.scatter(scaled_pct_pop_under_pov, scaled_violent_crimes_per_pop, alpha=0.5)
plt.title('Scatter Plot of Scaled Data')
plt.xlabel('Scaled PctPopUnderPov')
plt.ylabel('Scaled ViolentCrimesPerPop')
plt.show()
```
